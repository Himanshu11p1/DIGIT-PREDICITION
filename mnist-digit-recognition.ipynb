```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
from keras.datasets import mnist
#dataset=temsorflow.keras.datasets.mnist(train_images,train_labels),(test_images,test_labels)=dataset.load_data()
```

```python
(x_train,y_train),(x_test,y_test)=mnist.load_data()
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python

for i in range(9):
    plt.subplot(330+1+i)
    plt.imshow(x_train[i])
```

```python
for i in range(9):
    plt.subplot(330+1+i)
    plt.imshow(x_test[i])
```

```python
x_train=x_train/255
x_test=x_test/255
x_train
```

```python
x_test
```

```python
from keras.layers import Flatten,Dense,Dropout,Activation
from keras.models import Sequential
from keras.callbacks import EarlyStopping
from keras.layers.advanced_activations import LeakyReLU
model=Sequential([Flatten(input_shape=(28,28)),Dense(128,activation='relu'),Dense(100,activation='softmax'),Dense(30,activation='LeakyReLU')
                 ])
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(10,activation='relu'))
model.add(LeakyReLU(0.1))
model.add(Activation('softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
es=EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=3)
```

```python
history=model.fit(x_train,y_train,epochs=5,batch_size=64,validation_data=(x_test,y_test),callbacks=[es],
                 validation_split=0.25)
```

```python
model.evaluate(x_test,y_test)
```

```python
print(model.evaluate(x_train,y_train))
```

```python
plt.plot(model.evaluate(x_train,y_train))
plt.plot(model.evaluate(x_test,y_test))
plt.show()
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.show()
```

```python
prediction=model.predict(np.array([x_test[0]],dtype=float))
print("prediction: {}, label: {}".format(prediction,y_test[0]))
```

```python
predicted_result=model.predict(x_test)
class_preds=np.argmax(predicted_result,axis=-1)
from sklearn import metrics
p_score=metrics.precision_score(y_test,class_preds,average='macro')
print("Precision score is : ",p_score)
r_score=metrics.recall_score(y_test,class_preds,average='macro')
print("Recall score is : ",r_score)
c_r=metrics.classification_report(y_test,class_preds)
print("Classification report : ",c_r)
```
